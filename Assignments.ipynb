3.Write the end-to-end analytics workflow in the correct order and briefly explain each step.

Ans. 1.Collect the Data
This is where you actually pull the data together. Could be writing SQL queries, setting up API connections, or literally downloading CSV files. You're gathering all the raw ingredients.

2. Clean and Prepare the Data
Real data is messy as hell. You're dealing with missing values, duplicates, weird formatting, outliers that make no sense. This step takes forever but it's crucial - garbage in, garbage out, as they say.

3. Explore and Analyze
Time to dig in and see what the data is telling you. You're running statistical tests, calculating metrics, looking for patterns and correlations. This is where you start connecting dots and forming hypotheses.

4. Visualize the Results
Numbers in a spreadsheet don't convince anyone. You need to turn your findings into charts, dashboards, or graphs that actually make sense to people who aren't data nerds.

4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

Prompt engineering is basically the art and science of crafting instructions or questions to get the best possible output from AI models like ChatGPT, Claude, or other generative AI systems. Think of it like learning how to ask the right questions in the right way. The more specific and well-structured your prompt, the better the AI's response will be.

Two Common Types of Prompts:

1. Zero-Shot Prompts
This is when you ask the AI to do something without giving it any examples first. You just describe what you want and hope it figures it out.

2. Few-Shot Prompts
Here you give the AI a few examples of what you want before asking it to do the actual task.

5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles, responsibilities, and focus?

Ans.Business Analyst is more about the "why" and "what next." They're focused on understanding business problems, figuring out what the company needs, and bridging the gap between technical teams and stakeholders. They spend time in meetings, gathering requirements, documenting processes, and recommending solutions that might not even involve data sometimes. Think process improvement, stakeholder management, and strategic planning.
Data Analyst is all about the "what happened" and "what does it mean." They live in the data - pulling it, cleaning it, analyzing it, and creating reports or dashboards. Their job is to find patterns, trends, and insights buried in numbers. They're more technical with SQL, Python, Excel, and visualization tools. Less meetings, more spreadsheets. Business Analysts look at the bigger organizational picture and work on improving processes and systems. Data Analysts dive deep into datasets to extract meaningful insights that support decisions.

8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

Ans. 1. Descriptive Analytics - "What happened?"
Basic reporting and dashboards showing historical data. Sales numbers, traffic stats, etc.

2. Diagnostic Analytics - "Why did it happen?"
Digging into the data to find root causes and correlations behind the numbers.

3. Predictive Analytics - "What will happen?"
Using machine learning and statistical models to forecast future outcomes and trends.

4. Prescriptive Analytics - "What should we do?"
Goes beyond predictions to recommend specific actions and optimize decisions.
5. Cognitive Analytics - "Understanding like humans"

Processing unstructured data (text, images, voice) using NLP and AI to mimic human thinking.

6. Generative AI - "Create something new"
Creates entirely new content - text, images, code, videos - not just analyzes existing data.
